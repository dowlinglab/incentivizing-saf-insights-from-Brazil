```python
#Import necessary packages
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np
import matplotlib.hatch
from matplotlib.patches import Polygon
```

# Supply Chain Cost Sensitivty to SAF Blend Requirement and Decision-making Perspective: Recommended Cost Subsidies

```python
this_file_path = os.getcwd()

#Pull data for each case study

#Case 1
blend_mills1 = {}
for i in [0,10,20,30,40,50]:
    blend_mills1[i] = pd.read_csv(this_file_path + '\\Case1\\interest_mid_blend_' + str(i) + '\\key_results_mills.csv')
#1b
blend_mills2 = {} 
for i in [0,10,20,30,40,50]:
    blend_mills2[i] = pd.read_csv(this_file_path + '\\Case2\\interest_mid_blend_' + str(i) + '\\key_results_mills.csv')
#2a
blend_mills3 = {} 
for i in [0,10,20,30,40,50]:
    blend_mills3[i] = pd.read_csv(this_file_path + '\\Case3\\interest_mid_blend_' + str(i) + '\\key_results_mills.csv')
#2b
blend_mills4 = {} 
for i in [0,10,20,30,40,50]:
    blend_mills4[i] = pd.read_csv(this_file_path + '\\Case4\\interest_mid_blend_' + str(i) + '\\key_results_mills.csv')
```

```python
#Calculate the Additional Cost of SAF

total_jet = 4272564 #Total jet fuel demand in Brazil m3/year

#Case 1
additional1 = []
for i in [10,20,30,40,50]:
    additional1.append((blend_mills1[i]['sc cost'][9] - blend_mills1[0]['sc cost'][9])/(total_jet*(i/100))/1000)

#Case 2
additional2 = []
for i in [10,20,30,40,50]:
    additional2.append((blend_mills2[i]['sc cost'][9] - blend_mills2[0]['sc cost'][9])/(total_jet*(i/100))/1000)

#Case 3
additional3 = []
for i in [10,20,30,40,50]:
    additional3.append((blend_mills3[i]['sc cost'][9] - blend_mills3[0]['sc cost'][9])/(total_jet*(i/100))/1000)

#Case 4
additional4 = []
for i in [10,20,30,40,50]:
    additional4.append((blend_mills4[i]['sc cost'][9] - blend_mills4[0]['sc cost'][9])/(total_jet*(i/100))/1000)
```

```python
plt.plot([10,20,30,40,50],additional1, label = 'Case 1', color = 'red', marker='o')
plt.plot([10,20,30,40,50],additional2, label = 'Case 2',color = 'purple', marker='v')
plt.plot([10,20,30,40,50],additional3, label = 'Case 3', color = 'dodgerblue', marker='s')
plt.plot([10,20,30,40,50],additional4, label = 'Case 4', color = 'tab:brown', marker='^')
plt.ylim(top = 1.6, bottom = 1.05)
plt.ylabel('Additional Cost (R\$ L$^{-1}$)', fontweight = 'bold', fontsize = 14)
plt.xlabel('Blend Requirement (%)', fontweight = 'bold', fontsize = 14)
plt.text(22, 1.08, 'Upgrading at Mills', fontweight='bold', fontsize = 13)
plt.text(30,1.5,'Upgrading at Refineries', fontweight = 'bold', fontsize = 13)
plt.legend(fontsize = 14)
plt.xticks([10,20,30,40,50])

plt.savefig('Results_Figures/additionalSAFcost.png',bbox_inches='tight', dpi=500)
plt.show()
```
